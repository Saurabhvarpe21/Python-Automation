```python
folder_path = 'pan_data'
path = r'C:\Program Files (x86)\Tesseract-OCR\tesseract.exe'
image_list = os.listdir(folder_path)
pytesseract.pytesseract.tesseract_cmd = path


pan_num_list = []
dob_list = []
for image in image_list:
    image = os.path.join(folder_path, image)
    
    text = pytesseract.image_to_string(image)
    
    
    pan_number = re.findall(r'\b[A-Z]{8}\d{1}\b', text)
    dob = re.findall(r'\b\d{2}[/]\d{2}[/]\d{4}',text)
#     print(dob)

    if pan_number:
        pan_num_list.append(pan_number[0])
        
    else:
        pan_num_list.append('')
        
    if dob:
        dob_list.append(dob[0])
        
    else:
        dob_list.append('')
        
        
data = {'Image Names': image_list,
       'PAN No': pan_num_list,
       'DOB': dob_list}

data
```
